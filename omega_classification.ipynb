```python
# Covariance embeddings

# Open the covariances
features = h5io.read_hdf5(
            deriv_root / 'features_fb_covs_rest.h5')
subjects = list(features.keys())

# Alpha covariances of all control and parkinson subjects
covs = []
subjects_red = []
for sub in subjects:
    if features[sub]['covs'].shape == (7, 262, 262):
        covs.append(features[sub]['covs'])
        subjects_red.append(sub[4:])
covs = np.array(covs)

subjects_data_idx = subjects_data.set_index('subject_id')
groups = list(subjects_data_idx.loc[subjects_red]['group'].values)

print(covs.shape)
print(len(groups))
```

```python
class ProjCommonSpace(BaseEstimator, TransformerMixin):
    def __init__(self, scale=1, n_compo='full', reg=1e-7):
        self.scale_ = scale
        self.n_compo = n_compo
        self.reg = reg

    def fit(self, X, y=None):
        if self.scale_ == 'auto':
            self.scale_ = 1 / np.mean([np.trace(x) for x in X])
        self.n_compo = len(X[0]) if self.n_compo == 'full' else self.n_compo
        self.filters_ = []
        self.patterns_ = []
        C = X.mean(axis=0)
        eigvals, eigvecs = eigh(C)
        ix = np.argsort(np.abs(eigvals))[::-1]
        evecs = eigvecs[:, ix]
        evecs = evecs[:, :self.n_compo].T
        self.filters_.append(evecs)  # (fb, compo, chan) row vec
        self.patterns_.append(pinv(evecs).T)  # (fb, compo, chan)
        return self

    def transform(self, X):
        n_sub, _, _ = X.shape
        self.n_compo = len(X[0]) if self.n_compo == 'full' else self.n_compo
        Xout = np.empty((n_sub, self.n_compo, self.n_compo))
        Xs = self.scale_ * X
        filters = self.filters_[0]  # (compo, chan)
        for sub in range(n_sub):
            Xout[sub] = filters @ Xs[sub] @ filters.T
            Xout[sub] += self.reg * np.eye(self.n_compo)
        return Xout  # (sub , compo, compo)
```

```python
# Proj common space
rank = 150
band = 3

pcs = ProjCommonSpace(n_compo=rank)
covs_proj = pcs.fit_transform(covs [:, band])

# %% Spectral embedding
se = SpectralEmbedding()
covs_embedding = se.fit_transform(covs_proj)

print(covs_embedding.shape)
```

```python
# Plot embeddings
colors = [{'Control' : 'blue', 'Parkinson' : 'red', 'Chronic Pain' : 'yellow'}[i] for i in groups]
plt.figure(figsize=(8,8))

plt.scatter(covs_embedding[:,0], covs_embedding[:,1], c = colors)

blueC = mpatches.Circle((1,1), color='blue', label='Control')
redC = mpatches.Circle((1,1), color='red', label='Parkinson')
yellowC = mpatches.Circle((1,1), color='yellow', label='Chronic Pain')

plt.legend(handles = [blueC, redC, yellowC])
plt.title('Covariance embedding for each Omega subject')
plt.show()
```

```python
# PCA vecteur tangent
ts = TangentSpace()
covs_tangent = ts.fit_transform(covs_proj)

pca = PCA(n_components=2)
covs_pca = pca.fit_transform(covs_tangent)

# Dataframe PCA

df_pca_covs = pd.DataFrame({
        '1st compo': covs_pca[:, 0],
        '2nd compo': covs_pca[:, 1]
    })

df_pca_covs['group'] = groups

# Plot PCA

fig = plt.figure(figsize=(10,10))
ax = plt.subplot()
fig.add_subplot(ax)
ax.set_title('PCA on the tangent vectors of the covariances of Omega subjects')

sns.set_theme(style="whitegrid", font_scale=1.5)
sns.set_palette('colorblind')
sns.scatterplot(df_pca_covs, x = '1st compo', y = '2nd compo', hue = 'group', legend = 'full', ax = ax)
```

```python
# Classification code (from https://github.com/apmellot/Dementia_screening_challenge_biomag_2022/blob/main/final_model.py)

frequency_bands = {
    "low": (0.1, 1),
    "delta": (1, 4),
    "theta": (4.0, 8.0),
    "alpha": (8.0, 15.0),
    "beta_low": (15.0, 26.0),
    "beta_mid": (26.0, 35.0),
    "beta_high": (35.0, 49)
}

def get_subjects_labels(subjects):
    subjects_data_idx = subjects_data.set_index('subject_id')
    return list(subjects_data_idx.loc[subjects]['group'].values)


def get_subjects_age(age, labels):
    subjects = []
    participants_file = os.path.join(bids_root, "participants.tsv")
    subjects_info = pd.read_csv(participants_file, sep = '\t')
    for i in range(subjects_info.shape[0]):
        if subjects_info['participant_id'].iloc[i][4:] in list(subjects_data['subject_id'].values) and subjects_info['participant_id'].iloc[i] in good_sub:
            if subjects_info['age'].iloc[i] >= age and subjects_info['group'].iloc[i] in labels:
                subjects.append(subjects_info['participant_id'].iloc[i][4:])
    print(len(subjects))
    return subjects


def run_model():
    all_subjects = get_subjects_age(20, ['Control', 'Parkinson'])
    y = get_subjects_labels(all_subjects)

#    features_psd = h5io.read_hdf5(deriv_root / 'features_features_psd.h5')
#    X_psd_train = np.concatenate(
#        [features_psd[sub][None, :] for sub in train_subjects],
#        axis=0
#    )
#    X_psd_test = np.concatenate(
#        [features_psd[sub][None, :] for sub in test_subjects],
#        axis=0
#    )

    reg = 1e-7
    rank = 240
    features_covs = h5io.read_hdf5(deriv_root / 'features_fb_covs_rest.h5')

    covs = [features_covs['sub-'+sub]['covs'] for sub in all_subjects]
    covs = np.array(covs)
    X_covs = pd.DataFrame(
        {band: list(covs[:, ii]) for ii, band in
            enumerate(frequency_bands)})
    
    filter_bank_transformer = coffeine.make_filter_bank_transformer(
        names=list(frequency_bands),
        method='riemann',
        projection_params=dict(scale='auto', n_compo=rank, reg=reg)
    )

    X = filter_bank_transformer.fit_transform(X_covs)
#    X_train = np.concatenate((X_psd_train, X_covs_train), axis=1)
#    X_test = np.concatenate((X_psd_test, X_covs_test), axis=1)
    
    model = make_pipeline(
        StandardScaler(),
        LogisticRegressionCV(Cs=[0.1], max_iter=int(1e4))
    )

    metrics = [accuracy_score, balanced_accuracy_score]
    scoring = {m.__name__: make_scorer(m) for m in metrics}
    
    scores = cross_validate(model, X, y, scoring = scoring)

    return scores
```
