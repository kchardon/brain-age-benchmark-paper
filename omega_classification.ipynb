```python
import mne
import numpy as np
import pandas as pd
import pathlib
import os
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

bids_root = pathlib.Path('/storage/store2/data/Omega')
deriv_root = pathlib.Path('/storage/store3/work/kachardo/derivatives/omega')

participants_file = os.path.join(bids_root, "participants.tsv")
all_subjects = pd.read_csv(participants_file, sep = '\t') # All information about subjects
subjects_data = pd.read_csv('omega_subjects.csv') # Session and run used
```

```python
# Plot average psd for parkinson and control subjects

fig = plt.figure(figsize = (10,5))
ax = plt.subplot()
fig.add_subplot(ax)
i = 0
colors = {'Control' : 'blue', 'Parkinson' : 'red'}

for subject in os.listdir(deriv_root):
    if subject.startswith('sub'):
        id = subject[4:]
        group = subjects_data[subjects_data['subject_id']==id]['group'].iloc[0]
        if group == 'Control' or group == 'Parkinson' :
            session = subjects_data[subjects_data['subject_id']==id]['session'].iloc[0]
            epoch_file = os.path.join(deriv_root,subject, "ses-0"+str(session), "meg","sub-"+str(id)+"_ses-0"+str(session)+"_task-rest_proc-autoreject_epo.fif")
            epoch = mne.read_epochs(epoch_file)
            epoch.compute_psd(picks = 'meg').plot(color = colors[group], picks = 'meg', average = True,axes = ax, ci = None)
            i += 1

red_patch = mpatches.Patch(color='red', label='Parkinson')
blue_patch = mpatches.Patch(color='blue', label='Control')
plt.legend(handles=[red_patch, blue_patch])

print(i)        
ax.set_title('Average PSD for each Control and Parkinson subject after autoreject')
plt.show()
```

```python
# Plot average psd for parkinson and control group

group_psds = {'Control' : [], 'Parkinson' : []}

for subject in os.listdir(deriv_root):
    if subject.startswith('sub'):
        id = subject[4:]
        group = subjects_data[subjects_data['subject_id']==id]['group'].iloc[0]
        if group == 'Control' or group == 'Parkinson' :
            session = subjects_data[subjects_data['subject_id']==id]['session'].iloc[0]
            epoch_file = os.path.join(deriv_root,subject, "ses-0"+str(session), "meg","sub-"+str(id)+"_ses-0"+str(session)+"_task-rest_proc-autoreject_epo.fif")
            epoch = mne.read_epochs(epoch_file, verbose = False)
            spectrum, freqs = epoch.compute_psd(picks = 'meg').get_data(return_freqs=True)
            spectrum_avg = np.average(np.average(spectrum, axis=0), axis = 0)
            group_psds[group].append(spectrum_avg)
            i += 1
```

```python
control_psds = np.array(group_psds['Control'])
parkinson_psds = np.array(group_psds['Parkinson'])

control_mean = control_psds.mean(axis = 0)
control_std = control_psds.std(axis = 0)

parkinson_mean = parkinson_psds.mean(axis = 0)
parkinson_std = parkinson_psds.std(axis = 0)


fig = plt.figure(figsize = (10,5))
ax = plt.subplot()
fig.add_subplot(ax)
ax.set_yscale('log')
ax.set_xlim((0,60))

plt.plot(freqs, control_mean, 'k-', color = 'blue', label = 'Control')
plt.fill_between(freqs, control_mean - control_std, control_mean + control_std,  alpha=0.5, color = 'blue')

plt.plot(freqs, parkinson_mean, 'k-', color = 'red', label = 'Parkinson')
plt.fill_between(freqs, parkinson_mean - parkinson_std, parkinson_mean + parkinson_std,  alpha=0.5, color = 'red')

ax.set_title('Average PSD for each group')
plt.legend()

plt.show()
```
